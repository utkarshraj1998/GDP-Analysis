```python
# Removing the column="West Bengal1" as it has all null values
gdp_current_price=gdp_current_price.loc[:,~gdp_current_price.columns.isin(["West Bengal1"])]
gdp_current_price.isnull().sum(axis=1)
```

```python
# calculating the average growth of states over duartion 2013-14, 2014-15 and 2015-16 by "moving average"

m_avg=gdp_current_price.loc[:3,'Duration':]

#Dropping Union Territories
m_avg=m_avg.drop(['Andaman & Nicobar Islands','Chandigarh','Delhi','Puducherry'], axis=1)

m_avg=m_avg.iloc[:,1:].rolling(window=2).mean() #rolling average calculating

# Making appropriate transformation like creating index and sorting the values

m_avg['Duration']=['2013-14','2014-15','2015-16', '2016-17']
m_avg=m_avg.iloc[1:]
m_avg=m_avg.set_index('Duration',drop=True)
m_avg=m_avg.T
m_avg['Index']=range(len(m_avg))
m_avg['State']=m_avg.index
m_avg=m_avg.set_index('Index').sort_values(by='State')

m_avg.head(5)
```

```python
m_avg=m_avg[m_avg['2016-17'].isna()==False]
m_avg
```

```python
# Transfoming the dataframe into a catagorial view to plot

gsdp=m_avg.append(m_avg.iloc[0:,1:],sort=False)
gsdp.loc[np.isnan(gsdp['2014-15']),['2014-15']]=m_avg['2015-16']
gsdp=gsdp.append(gsdp.iloc[0:,1:],sort=False)
gsdp.loc[np.isnan(gsdp['2014-15']),['2014-15']]=m_avg['2016-17']
gsdp=gsdp.iloc[0:,0::3]

gsdp.columns=['Rolling_Average','State'] # Renaming the column names
gsdp.drop_duplicates(subset = None, keep = 'first', inplace = True)  # removing duplicates

# Updating the durations

gsdp['Duration']=['2014-15' for i in range(len(gsdp))]
gsdp.iloc[9:,2]='2015-16'
gsdp.iloc[18:,2]='2016-17'

gsdp['Rolling_Average']=gsdp['Rolling_Average'].apply(lambda x:round(x,2)) # rounding the decimal variables in 2
gsdp.sort_values(by=['Duration','Rolling_Average'],ascending=True,inplace=True) # sorting

gsdp.head()
```

```python

# Now Plotting

# Setting gridlines,context and size of the figure

sns.set_style('whitegrid')
sns.set_context('talk')
plt.figure(figsize=(20,10))
plt.rcParams["axes.labelsize"] = 20
plt.rcParams['axes.titlesize'] = 22
plt.rcParams['axes.titlepad'] = 30

# Using barplot for categorical visualisation

sns.barplot(x='State',y='Rolling_Average',hue='Duration',data=gsdp,palette='tab10')
plt.xticks(rotation=90)

# Setting labels for x-axis, y-axis and title

plt.xlabel('State')
plt.ylabel('Average Growth(%)')
plt.title("(%)Average growth of states"' '"(2014-15, 2015-16 and 2016-17)")

# Executing

plt.show()
```

```python
gsva_secondary=gsva_drop[["Per Capita GSDP (Rs.)","Upper Primary - 2014-2015"]]
gsva_secondary.dropna(inplace=True)
gsva_secondary['Upper Primary - 2014-2015']=gsva_secondary['Upper Primary - 2014-2015'].astype(int)
gsva_secondary.plot.scatter(x="Per Capita GSDP (Rs.)",y="Upper Primary - 2014-2015")
plt.show()
```

```python
gsva_secondary=gsva_drop[["Per Capita GSDP (Rs.)","Secondary - 2014-2015"]]
gsva_secondary.dropna(inplace=True)
gsva_secondary['Secondary - 2014-2015']=gsva_secondary['Secondary - 2014-2015'].astype(int)
gsva_secondary.plot.scatter(x="Per Capita GSDP (Rs.)",y="Secondary - 2014-2015")
plt.show()
```

### Insights from Correlation of GDP per capita with dropout rates in education (primary, upper primary and secondary) for the year 2014-2015

* From the above plots we can tell that the drop rates is inversely proportional to the Per capita GSDP. So, if the drop rates are high, per capita GSDP will be low and if drop rates are low, per capita GSDP will be high.
* Drop out rates are higher in Secondary Education.
* Drop out rates for Primary and Upper Primary are less than or equal to 15.
* Drop out rates for Secondary is generally above 5.
* Statistics for Primary and Upper Primary are pretty similar.
* The outliers are due to Goa and Sikkim because of its Per Capita GDP.
#### Hence, we can conclude that Drop out rates does affect the per capita GSDP if it is not taken care of.
